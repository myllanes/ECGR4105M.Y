```python
#Michael Yllanes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/myllanes/ECGR4105M.Y./main/D3.csv', na_values='*')
```

# Problem 1 X1

```python
m = data.X1.size
X_0 = np.ones((m, 1))
X_0 = pd.DataFrame(X_0)
data.insert(0, "intercept",X_0)
display(data)
```

```python
CX1 = data.loc[:,['intercept','X1']]
CY1 = data.Y
theta1 = np.zeros(2)
iterations1 = 500;
alpha1 = 0.1;
```

```python
def computeCost(X,y,theta):
  m = len(X)
  predictions = X.dot(theta)
  errors = np.subtract(predictions, y)
  sqrErrors = np.square(errors)
  J = 1 / (2 * m) * np.sum(sqrErrors)
  return J
```

```python
def gradientDescent(X, y, theta, alpha, iterations):
  cost_history = np.zeros(iterations)

  for i in range(iterations):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sum_delta = (alpha / m) * X.transpose().dot(errors);
    theta = theta - sum_delta;
    cost_history[i] = computeCost(X, y, theta)
  return theta, cost_history
```

```python
Xarray1 = CX1.loc[:,['intercept','X1']].values
Yarray1 = CY1.values
theta1, cost_history1 = gradientDescent(Xarray1, Yarray1, theta1, alpha1, iterations1)
print('X1 Final value of theta =', theta1)
#print('X1 cost_history =', cost_history1)
```

```python
# Column of index 1 only
plt.scatter(CX1.X1, CY1, color='red', marker= '^', label= 'Training Data')
plt.plot(CX1.X1, CX1.dot(theta1), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()
```

```python
plt.plot(range(1, iterations1 + 1),cost_history1, color='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')
```

```python
# Grid to calculate J for X1
theta0_vals1 = np.linspace(4, 8, 500);
theta1_vals1 = np.linspace(-3, -1, 100);
# Set J vals to a matrix
J_vals_1 = np.zeros(shape = (len(theta1_vals1),len(theta0_vals1)));
for i in range(len(theta0_vals1)):
  for j in range(len(theta1_vals1)):
    t = [theta0_vals1[i], theta1_vals1[j]]
    J_vals_1[j,i] = computeCost(CX1, CY1, t)
```

```python
#print (J_vals_1)
```

```python
import matplotlib.pyplot as plt
# Plot J_vals
plt.contour(theta0_vals1, theta1_vals1, np.log(J_vals_1))
plt.xlabel('$theta_0$')
plt.ylabel('$theta_1$')
plt.scatter(theta1[0],theta1[1])
plt.grid()
plt.show()
```

# Problem 2

```python
# Define the compute_cost function
def compute_cost2(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1/(2*m)) * np.sum(np.square(predictions - y))
    return cost

# Define the gradient_descent function
def gradient_descent2(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (1/m) * X.T.dot(errors)
        theta -= alpha * gradient
        cost_history[i] = compute_cost2(X, y, theta)

    return theta, cost_history
```

```python
CX4 = data.loc[:,['intercept','X1','X2','X3']]
CY4 = data.Y

iterations4 = 3000;
alpha4 = 0.01;
```

```python
Xarray4 = CX4.loc[:,['intercept','X1','X2','X3']].values
Yarray4 = CY4.values
theta4 = np.zeros(Xarray4.shape[1])
theta4, cost_history4 = gradient_descent2(Xarray4, Yarray4, theta4, alpha4, iterations4)
print('X1 Final value of theta =', theta4)
print('X1 cost_history =', cost_history4)
```

```python
plt.scatter(Xarray4[:, 1], Yarray4, color='blue', label='X1')  # X1
plt.scatter(Xarray4[:, 2], Yarray4, color='red', label='X2')   # X2
plt.scatter(Xarray4[:, 3], Yarray4, color='black', label='X3') # X3

plt.plot(Xarray4[:, 1], CX4.dot(theta4), color='green', label='Regression Line')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()
```
